## K Nearest Neighbor from scratch in Python using Iris dataset

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/home/atul/Documents/BDA 2nd sem/ml/ml practice/LAB TASK/iris.data',header=None)
train,test=train_test_split(df,test_size=0.2)
train
```

```python
x=train.iloc[:,:-1].values
x
```

### train's target values

```python
y=train.iloc[:,-1:].values
y
```

```python
x[0:1,:]
```

```python
def knn(tdp,x,k):
    d=[]
    out=[]
    for i in range(len(x)):
        d.append(np.sqrt(np.linalg.norm(x[i:i+1,:]-tdp.values))) # Calculatong distance b/w test data point and all the train data points
    indx=np.argsort(d) #index of sorted d in ascending order
    for i in range(k):
        out.append(y[indx[i]])  #picking k data points' class at smallest distance from test data point        
    return(np.array(out))
```

```python
def predictY(out_knn,k):
    count_s=count_vc=count_vg=0 #To predict class on the basis of max votes for a test data point
    for i in range(len(out_knn)):
        if(out_knn[i][0]=='Iris-versicolor'):
            count_vc=count_vc+1
        if(out_knn[i][0]=='Iris-setosa'):
            count_s=count_s+1
        if(out_knn[i][0]=='Iris-virginica'):
            count_vg=count_vg+1
    count=np.array([count_vc,count_s,count_vg])
    if(np.argmax(count)==0): #index of class having maximum votes
        pred='Iris-versicolor'
    elif(np.argmax(count)==1):
        pred='Iris-setosa'
    else:
        pred='Iris-virginica'
    return(pred)
```

```python
def finalPrediction(x,test,k): #To predict classes for whole test dataset
    finalPredict=[]
    for i in range(len(test)):
        out_knn=knn(test.iloc[i:i+1,:-1],x,k)
        finalPredict.append(predictY(out_knn,k))
    return(finalPredict)
```

```python
finalPredict=finalPrediction(x,test,10)
finalPredict
```

```python
ytest=test.iloc[:,-1:].values #True test target values
ytest
```

## Testing Classification Accuracy
